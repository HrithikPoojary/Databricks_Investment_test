```python
%sql
select exists(
select * from ashlamba.bronze.rls_mapping
where user_id = current_user()
and region = 'south') validornot;
```

```python
%sql
create or replace function ashlamba.bronze.rls_region_restriction(p_region string)
return exists(
    select * from ashlamba.bronze.rls_mapping
    where user_id = current_user()
    and region = lower(p_region)
);
```

```python
%sql
alter table ashlamba.silver.stores_enr
set row filter ashlamba.bronze.rls_region_restriction on (region)
```
