```python
%sql
create or replace table ashlamba.default.employees(
    id int,
    name string ,
    salary int 
)
tblproperties(
    delta.enableChangeDataFeed = True
)
```

```python
%sql
alter table ashlamba.default.employees 
set tblproperties(
    delta.enableChangeDataFeed = True
)
```

```python
%sql
insert into ashlamba.default.employees
values (1 , 'John' , 50000),
       (2 , 'Alice' , 60000);
```
