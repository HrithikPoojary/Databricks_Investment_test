```sql
with join_sum_query 
as (
    select  p.product_name , round(sum(s.total_amount),2) as total_amount from sales_enr s
    left join product_enr p
    on s.product_id = p.product_id
    group by p.product_name),
ranked_query 
as (
    select * , dense_rank() over(order by total_amount desc) as rnk 
    from join_sum_query
)
select * from ranked_query
where rnk <= :top_product
;
```
