```python
data = [(1,100) , (2,200) , (3,300)]

schema = 'id int , salary int'

df = spark.createDataFrame(data = data , schema = schema)

df.display()
```

```python
df.write.format("delta")\
            .mode("append")\
                .save("/Volumes/ashlamba/bronze/bronze_volume/DeltaOptimize/")
```
