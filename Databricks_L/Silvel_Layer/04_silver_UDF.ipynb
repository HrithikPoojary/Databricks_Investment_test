```python
udf_greet = udf(greet_udf)
```

```python
df = spark.read.table("ashlamba.silver.product_enr")
```

```python
df.withColumn("greet_product" , udf_greet("category")).display()
```

```python
from pyspark.sql.functions import udf,col
from pyspark.sql.types import StringType

@udf(returnType= StringType())
def upper_letter(name):
    return "Upper Hello" + str(name)

df.select(upper_letter(col("category"))).display()
```

```python
def geeting_hello(name):
    return 'Hello ' + name
```

```python
spark.udf.register("greeting" , geeting_hello)
```

```python
spark.sql('select greeting(category) from ashlamba.silver.product_enr').display()
```
