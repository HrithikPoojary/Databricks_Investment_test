```python
from pyspark.sql.functions import rand,floor,expr ,col,lit 

df = spark.createDataFrame(data = [("a",100),("b",222)] , schema = 'name string , id int')

df.select( floor(lit(rand() * col('id'))).alias("num") ).show()
```

```python
from pyspark.sql.functions import floor,rand ,concat,col ,lit

(
    df.withColumn("salt_key" , floor(rand()*100))
            .withColumn("salt_key_id" , concat(col("id") , lit('_') , col("salt_key") ) )
            .show(truncate = False)
 
 )
```
